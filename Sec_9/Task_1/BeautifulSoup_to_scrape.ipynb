```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://quotes.toscrape.com/"
res=requests.get(url)
if(res.status_code==200):
  print("opened")
```

```python
soup=BeautifulSoup(res.content,"html.parser")
```

```python
from ast import keyword
quote=soup.find_all("span",class_="text")
df=[]
for quote in quote:
  text=quote.text
  auther=quote.find_next("small",class_="author")
  keyword=quote.find_next("div",class_="tags")
  df.append({"text":text,"auther":auther.text,"keyword":keyword.text})
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(df)
```

```python
import requests
from bs4 import BeautifulSoup
import time

data = []
i = 1
while True:
    url = f"https://quotes.toscrape.com/page/{i}/"
    res = requests.get(url)

    if res.status_code == 200:
        soup = BeautifulSoup(res.text, "html.parser")
        quotes = soup.find_all("div", class_="quote")

        for q in quotes:
            text = q.find("span", class_="text").text
            author = q.find("small", class_="author").text
            tags = [tag.text for tag in q.find_all("a", class_="tag")]

            data.append({
                "text": text,
                "author": author,
                "keywords": tags
            })
        i += 1
        time.sleep(1)
    else:
        print("No more pages...")
        break

# طباعة أول 5 عناصر كمثال
for d in data[:5]:
    print(d)
```

```python
import requests
from bs4 import BeautifulSoup
import time

data = []
i = 1

while True:
    url = f"https://quotes.toscrape.com/page/{i}/"
    res = requests.get(url)

    if res.status_code == 200:
        soup = BeautifulSoup(res.text, "html.parser")
        quotes = soup.find_all("div", class_="quote")


        if not quotes:
            print("No more quotes...")
            break

        for q in quotes:
            text = q.find("span", class_="text").text
            author = q.find("small", class_="author").text
            tags = [tag.text for tag in q.find_all("a", class_="tag")]

            data.append({
                "text": text,
                "author": author,
                "keywords": tags
            })

        i += 1
        time.sleep(0.5)
    else:
        print("Error fetching page...")
        break
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import csv

dri = webdriver.Chrome()
dri.get("https://quotes.toscrape.com")

quotes_list = []


while True:
    quotes = dri.find_elements(By.CLASS_NAME, "quote")
    for q in quotes:
        text = q.find_element(By.CLASS_NAME, "text").text
        author = q.find_element(By.CLASS_NAME, "author").text
        tags = [tag.text for tag in q.find_elements(By.CLASS_NAME, "tag")]

        quotes_list.append({
            "text": text,
            "author": author,
            "tags": tags
        })
    try:
        go_next = dri.find_element(By.LINK_TEXT, "Next →")
        go_next.click()
        time.sleep(1)
    except:
        break
dri.quit()


with open("quotes.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["text", "author", "tags"])
    writer.writeheader()
    writer.writerows(quotes_list)
print(f"extracted{len(quotes_list)}")


```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("muhammedfathi/arabic-news-texts-corpus")

print("Path to dataset files:", path)
```

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import re
```

```python
base_path = "/root/.cache/kagglehub/datasets/muhammedfathi/arabic-news-texts-corpus/versions/1"
file_path = os.path.join(base_path, "arabic_categorization_data.csv")
```

```python
df = pd.read_csv(os.path.join(path, "arabic_categorization_data.csv"))
df.head()
```

```python
df['type'].value_counts()
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
df.head()
```

```python
def preprocess_arabic(text):
    text = str(text)
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\d+', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

df["clean_text"] = df["text"].apply(preprocess_arabic)
```

```python
df.head()
```

```python
df.drop(columns=["text"], inplace=True)
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df["clean_text"], df["type"], test_size=0.2, random_state=42, stratify=df["type"])
```

```python
print('Training Data Shape:', X_train.shape)
print('Testing Data Shape: ', X_test.shape)
```

```python
train_data = pd.DataFrame(X_train)
train_data['type'] = y_train

# Now train_data contains both features and the target
train_data.head()
```

```python
train_data.to_csv('train_data.csv')
```

```python
train_data
```

```python
test_data = pd.DataFrame(X_test)
test_data['type'] = y_test

# Now train_data contains both features and the target
test_data.head()
```

```python
test_data.to_csv('test_data.csv')
```

```python
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
y_train.value_counts()
```

```python
clf = LogisticRegression(max_iter=200)
clf.fit(X_train_vec, y_train)
```

```python
y_pred = clf.predict(X_test_vec)
print("\n✅ Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

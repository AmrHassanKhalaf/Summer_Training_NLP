```python
import pandas as pd
from sklearn.metrics import classification_report, accuracy_score
```

```python
test=pd.read_csv('/content/test_data.csv')
```

```python
test
```

```python
test.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
# Drop rows with NaN values in the 'clean_text' column
test.dropna(subset=['clean_text'], inplace=True)
x = test['clean_text']
y = test['target']
```

```python
import pickle
with open(r"/content/text_clf_model.pkl", 'rb') as f:
 model = pickle.load(f)
```

```python
def predict(x,model):
    pred=model.predict(x)
    return pred
```

```python
predictions_vc_clf = []

for index in range(len(x)):

    row_text = x.iloc[[index]]

    print(row_text)
    predictions_vc_clf.append(predict(row_text, model)[0])
```

```python
from sklearn import metrics

print(metrics.confusion_matrix(y,predictions_vc_clf))

print(metrics.classification_report(y,predictions_vc_clf))

print(metrics.accuracy_score(y,predictions_vc_clf))
```

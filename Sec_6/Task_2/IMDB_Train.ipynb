```python
from tensorflow.keras.datasets import imdb
```

```python
# Load data (only keep top 10,000 words)
(x_train, y_train), (_, _) = imdb.load_data(num_words=10000)
```

```python
print(x_train[0])
print(y_train[0])
```

```python
from tensorflow.keras.preprocessing import sequence
num_words=10000
maxlen=250
x_train = sequence.pad_sequences(x_train, maxlen=maxlen)
```

```python
x_train.shape
```

```python
import numpy as np

def to_bow(sequences, num_words):
    bow = np.zeros((len(sequences), num_words))
    for i, seq in enumerate(sequences):
        for word_id in seq:
            if word_id < num_words:
                bow[i, word_id] = 1
    return bow

x_train_bow = to_bow(x_train, num_words)

print(x_train_bow.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python


clf = LogisticRegression(max_iter=200)
clf.fit(x_train_bow, y_train)
```

```python
import pickle

with open("clf.pkl", "wb") as f:
    pickle.dump(clf, f)
```

```python
from sklearn.svm import LinearSVC
sv=LinearSVC()
sv.fit(x_train_bow, y_train)
```

```python
import pickle

with open("sv.pkl","wb") as a:
  pickle.dump(sv,a)
```
